```python
import numpy as np
import os 
import sys
import pandas as pd
from keras.models import Sequential,Model
from keras.layers import *
from keras.datasets import mnist
from keras.optimizers import Adam
from keras.layers.advanced_activations import LeakyReLU
from os import listdir
from skimage import io
from scipy.misc import imresize
from PIL import Image
import matplotlib.pyplot as plt
import math
```

```python
def generator():
    model = Sequential()
    
    model.add(Dense(4*4*1024,input_dim = 100 , activation = 'relu'))
    model.add(BatchNormalization())
    model.add(Reshape((4,4,1024)))
    # 4x4x1024
    
    model.add(Conv2DTranspose(512, (5,5), padding = 'valid', activation = 'relu'))
    model.add(BatchNormalization())
    # 8x8x512
    
    model.add(Conv2DTranspose(256, (5,5), padding = 'same', activation = 'relu', strides=(2,2)))
    model.add(BatchNormalization())
    # 16x16x256
    
    
    model.add(Conv2DTranspose(128, (5,5), padding = 'same', activation = 'relu', strides=(2,2)))
    model.add(BatchNormalization())
    # 32x32x128
    
    model.add(Conv2DTranspose(64, (5,5), padding = 'same', activation = 'relu',  strides=(2,2)))
    model.add(BatchNormalization())
    # 64x64x64
    
    model.add(Conv2DTranspose(32, (5,5), padding = 'same', activation = 'relu',  strides=(2,2)))
    model.add(BatchNormalization())
    # 128x128x32
    
    model.add(Conv2DTranspose(1, (5,5), padding = 'same', activation = 'tanh',  strides=(2,2)))
    # 256x256x1

    return model

gen_model = generator()
print(gen_model.summary())
```

```python
def discriminator():
    model = Sequential()
    model.add(Conv2D(32,(5,5),padding = 'same',input_shape = (256,256,1), strides=(2,2)))
    model.add(LeakyReLU(0.2))
    # 128x128x32
    
    model.add(Conv2D(64,(5,5),padding = 'same',  strides=(2,2)))
    model.add(LeakyReLU(0.2))
    model.add(BatchNormalization())
    # 64x64x64
    
    model.add(Conv2D(128,(5,5),padding = 'same',  strides=(2,2)))
    model.add(LeakyReLU(0.2))
    model.add(BatchNormalization())
    # 32x32x128
    
    model.add(Conv2D(256,(5,5),padding = 'same',  strides=(2,2)))
    model.add(LeakyReLU(0.2))
    model.add(BatchNormalization())
    #16x16x256
    
    model.add(Conv2D(512,(5,5),padding = 'same', strides=(2,2)))
    model.add(LeakyReLU(0.2))
    model.add(BatchNormalization())
    #8x8x512
    
    model.add(Conv2D(1024,(5,5),padding = 'same', strides=(2,2)))
    model.add(LeakyReLU(0.2))
    model.add(BatchNormalization())
    #4x4x1024
    
    model.add(Flatten())
    model.add(BatchNormalization())
    model.add(Dense(1, activation='sigmoid'))
    
    
    return model

disc_model = discriminator() 
disc_model.summary()
```

```python
optimizer = Adam(lr=0.0002, beta_1=0.5)
disc_model.compile(loss = 'binary_crossentropy' , optimizer = optimizer )
gen_model.compile(loss = 'binary_crossentropy' , optimizer = optimizer )
```

```python
#Build Adverserial Model using gen and disc 
gen_inp = Input(shape = (100,))

gen_out = gen_model(gen_inp)

disc_model.trainable = False

disc_out = disc_model(gen_out)

adv_model = Model(input = gen_inp , output = disc_out)

adv_model.summary()
adv_model.compile(loss = 'binary_crossentropy' , optimizer = optimizer )
```

```python
def get_img(data_path):
    # Getting image array from path:
    img_size = 256
    img = io.imread(data_path)
    img = Image.fromarray(img)
    img = np.array(img.resize((img_size,img_size)))
    return img

def get_dataset(dataset_path="./no_background"):
    img_size = 256
    # Getting all data from data path:
    try:
        X = np.load('Data/npy_train_data/X.npy')
        Y = np.load('Data/npy_train_data/Y.npy')
    except:
        labels = listdir(dataset_path) # Geting labels
        print('Categories:\n', labels)
        len_datas = 0
        for label in labels:
            len_datas += len(listdir(dataset_path+'/'+label))
        X = np.zeros((len_datas, img_size, img_size), dtype='float32')
        Y = np.zeros(len_datas)
        count_data = 0
        count_categori = [-1,''] # For encode labels
        for label in labels:
            datas_path = dataset_path+'/'+label
            for data in listdir(datas_path):
                #print("Imagen cargada"+data+label)
                img = get_img(datas_path+'/'+data)
                X[count_data] = img
                # For encode labels:
                if label != count_categori[1]:
                    count_categori[0] += 1
                    count_categori[1] = label
                Y[count_data] = count_categori[0]
                count_data += 1
        # Create dateset:        
        #Y = keras.utils.to_categorical(Y)
        if not os.path.exists('Data/npy_train_data/'):
               os.makedirs('Data/npy_train_data/')
        np.save('Data/npy_train_data/X.npy', X)
        np.save('Data/npy_train_data/Y.npy', Y)
    #X /= 255.

    return X, Y

X, y =get_dataset()
```

```python
def generator_containing_discriminator(g, d):
    model = Sequential()
    model.add(g)
    d.trainable = False
    model.add(d)
    return model
```

```python
def combine_images(generated_images):
    num = generated_images.shape[0]
    width = int(math.sqrt(num))
    height = int(math.ceil(float(num)/width))
    shape = generated_images.shape[1:3]
    image = np.zeros((height*shape[0], width*shape[1]),
                     dtype=generated_images.dtype)
    for index, img in enumerate(generated_images):
        i = int(index/width)
        j = index % width
        image[i*shape[0]:(i+1)*shape[0], j*shape[1]:(j+1)*shape[1]] = \
            img[:, :, 0]
    return image
```

```python
def train( X_train, epochs, BATCH_SIZE=128):
    d = discriminator()
    g = generator()
    d_on_g = generator_containing_discriminator(g, d)
    d_optim = Adam(lr=0.0002, beta_1=0.5)
    g_optim = Adam(lr=0.0002, beta_1=0.5)
    g.compile(loss='binary_crossentropy', optimizer=g_optim)
    d_on_g.compile(loss='binary_crossentropy', optimizer=g_optim)
    d.trainable = True
    d.compile(loss='binary_crossentropy', optimizer=d_optim)
    for epoch in range(epochs):
        print("Epoch is", epoch)
        print("Number of batches", int(X_train.shape[0]/BATCH_SIZE))
        for index in range(int(X_train.shape[0]/BATCH_SIZE)):
            noise = np.random.uniform(-1, 1, size=(BATCH_SIZE, 100))
            image_batch_aux = X_train[index*BATCH_SIZE:(index+1)*BATCH_SIZE]
            image_batch=np.reshape(image_batch_aux, (image_batch_aux.shape[0],image_batch_aux.shape[1],image_batch_aux.shape[2],-1))
            generated_images = g.predict(noise, verbose=0)
            
            
            if index % 20 == 0:
                image = combine_images(generated_images)
                plt.imshow(image)
                #image = image*127.5+127.5
                #plt.imsave(arr= image, cmap='Greys_r', fname=str(epoch)+"_"+str(index)+".jpeg")
#                 Image.fromarray(image.astype(np.float32)).save(
#                     str(epoch)+"_"+str(index)+".jpeg")
            X = np.concatenate((image_batch, generated_images))
            y = [1] * BATCH_SIZE + [0] * BATCH_SIZE
            d_loss = d.train_on_batch(X, y)
            print("batch %d d_loss : %f" % (index, d_loss))
            noise = np.random.uniform(-1, 1, (BATCH_SIZE, 100))
            d.trainable = False
            g_loss = d_on_g.train_on_batch(noise, [1] * BATCH_SIZE)
            d.trainable = True
            print("batch %d g_loss : %f" % (index, g_loss))
            if index % 10 == 9:
                g.save_weights('generator', True)
                d.save_weights('discriminator', True)
```

```python
train(X_train, 100)
```
